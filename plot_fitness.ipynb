#### Run the cells below to visualize the fitness trajectory of each robot during training.

```python
import numpy as np
from matplotlib import pyplot as plt
plt.rcParams['figure.dpi'] = 150
```

```python
fitness_history = np.load("fitness_history.npy")
n_robots = fitness_history.shape[0]

fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(16, 16), sharex=True, sharey=True)
fig.suptitle('Robot Fitness Trajectory', fontsize=18, fontweight='bold', y=0.995)

axes_flat = axes.flatten() if n_robots > 1 else [axes]

for idx in range(n_robots):
    ax = axes_flat[idx]
    
    steps = np.arange(len(fitness_history[idx]))
    ax.plot(steps, fitness_history[idx], linewidth=2, color='#2E86AB', alpha=0.8)
    
    final_fitness = fitness_history[idx][-1]
    ax.scatter(len(fitness_history[idx])-1, final_fitness, 
               color='#E63946', s=80, zorder=5, edgecolors='white', linewidth=1.5)
    
    ax.set_title(f'Robot {idx} (Final: {final_fitness:.2f})', fontsize=12, fontweight='bold')
    ax.set_xlabel('Training Step', fontsize=10)
    if idx % 4 == 0:
        ax.set_ylabel('Fitness (distance travelled)', fontsize=10)
    ax.grid(True, alpha=0.3, linestyle='--')
    ax.set_facecolor('#f8f9fa')


plt.tight_layout()
plt.show()
```
